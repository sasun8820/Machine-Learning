```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("mtcars.csv")
```

#### 2.  Display the first five rows of the data.

```python
data.head(5)
```

#### 3.  Calculate the average of the mpg column for all cars within each category of the cyl column.

```python
grouped = data.groupby(['cyl'])
avg_mpg = grouped.mean(numeric_only=True)['mpg']
avg_mpg
```

#### 4.  Create a histogram using the mpg column

The below histogram of mpg levels indicates that the trend is right-skewed, with the maximum frequency of mpg levels centered around 15-20 and only a few at higher levels.

```python
plt.figure(figsize=(6, 6))
plt.hist(data['mpg'],color='skyblue', edgecolor='black')
plt.title('Histogram: Test Score')
plt.xlabel('Test Score')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### 5. Choose two variables in the data and create a scatterplot.

I have made the scatterplot of mpg level and car weight to see how those two variables might be correlated. From the general declining trend, we could infer that the MPG level (miles per gallon) seems to be higher in cars with lower weights, which makes intuitive sense given that heavier cars may consumer more gas per gallon, causing their MPG level to be lower than that of lighter weights.

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['mpg'], data['wt'], c='blue', marker='o', edgecolors='black')
plt.title('Scatterplot of MPG vs. Weight')
plt.xlabel('MPG')
plt.ylabel('Weight (1000 lbs)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
